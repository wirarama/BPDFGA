```python
board = []
for i in range(8):
    row = [i for i in range(8)]
    board.append(row)
    
board = [[(i+1)*(j+1) for i in range(8)] for j in range(8)]
print(board,len(board))
```

```python
EMPTY = "-"
ROOK = "ROOK"
board = []
for i in range(8):
    row = [EMPTY for i in range(8)]
    board.append(row)
board[0][0] = ROOK
board[0][7] = ROOK
board[7][0] = ROOK
board[7][7] = ROOK
for x in board:
    print(x)
```
